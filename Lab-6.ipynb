```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate and visualize normal distribution
data = np.random.normal(loc=60, scale=12, size=1000)

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Plot Probability Density Function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, 60, 12)
plt.plot(x, p, 'k', linewidth=2)

plt.title('Histogram and PDF of Normal Distribution (Mean=60, Std=12)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Probability calculations for calorie intake
mean_cal = 30
std_cal = 5

# Probabilities using CDF
greater_than_19 = 1 - stats.norm.cdf(19, mean_cal, std_cal)
less_than_19 = stats.norm.cdf(19, mean_cal, std_cal)
between_19_and_35 = stats.norm.cdf(35, mean_cal, std_cal) - stats.norm.cdf(19, mean_cal, std_cal)

print(f"P(X >= 19): {greater_than_19:.4f}")
print(f"P(X <= 19): {less_than_19:.4f}")
print(f"P(19 <= X <= 35): {between_19_and_35:.4f}")

# Visualization of probabilities
x = np.linspace(10, 50, 1000)
y = stats.norm.pdf(x, mean_cal, std_cal)

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b', label='Normal Distribution')
plt.fill_between(x, y, where=(x >= 19), color='red', alpha=0.3, label='X >= 19')
plt.fill_between(x, y, where=(x <= 19), color='green', alpha=0.3, label='X <= 19')
plt.fill_between(x, y, where=((x >= 19) & (x <= 35)), color='purple', alpha=0.3, label='19 <= X <= 35')

plt.title('Probability Visualization for Daily Calorie Intake')
plt.xlabel('Calorie Intake')
plt.ylabel('Density')
plt.legend()
plt.show()
```
