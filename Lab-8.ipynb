```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Data
daily_calorie_intake = np.array([2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000])
exercise_duration = np.array([60, 50, 30, 90, 45, 20, 75, 95, 15, 10])
bmi = np.array([25, 26, 28, 23, 27, 30, 24, 22, 31, 33])

# Scatter Plot: Daily Calorie Intake vs. BMI
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.scatter(daily_calorie_intake, bmi, color='b')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('BMI')
plt.title('Scatter Plot: Daily Calorie Intake vs. BMI')
plt.grid(True)

# Scatter Plot: Exercise Duration vs. BMI
plt.subplot(3, 1, 2)
plt.scatter(exercise_duration, bmi, color='r')
plt.xlabel('Exercise Duration (minutes)')
plt.ylabel('BMI')
plt.title('Scatter Plot: Exercise Duration vs. BMI')
plt.grid(True)

# Scatter Plot: Daily Calorie Intake vs. Exercise Duration
plt.subplot(3, 1, 3)
plt.scatter(daily_calorie_intake, exercise_duration, color='g')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('Exercise Duration (minutes)')
plt.title('Scatter Plot: Daily Calorie Intake vs. Exercise Duration')
plt.grid(True)

plt.tight_layout()
plt.show()

# Calculate Pearson correlation coefficients
correlation_calories_bmi, _ = pearsonr(daily_calorie_intake, bmi)
correlation_exercise_bmi, _ = pearsonr(exercise_duration, bmi)
correlation_calories_exercise, _ = pearsonr(daily_calorie_intake, exercise_duration)

# Display Pearson correlation results
print(f"Pearson correlation coefficient (Calories vs BMI): {correlation_calories_bmi:.2f}")
print(f"Pearson correlation coefficient (Exercise vs BMI): {correlation_exercise_bmi:.2f}")
print(f"Pearson correlation coefficient (Calories vs Exercise Duration): {correlation_calories_exercise:.2f}")

# Interpret the results
if correlation_calories_bmi > 0:
    print("There is a positive correlation between Daily Calorie Intake and BMI, indicating that higher calorie intake tends to be associated with a higher BMI.")
else:
    print("There is no significant positive correlation between Daily Calorie Intake and BMI.")

if correlation_exercise_bmi < 0:
    print("There is a negative correlation between Exercise Duration and BMI, indicating that longer exercise durations are associated with lower BMI.")
else:
    print("There is no significant negative correlation between Exercise Duration and BMI.")
```

```python
import numpy as np

# Data
daily_calorie_intake = np.array([2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000])
exercise_duration = np.array([60, 50, 30, 90, 45, 20, 75, 95, 15, 10])
bmi = np.array([25, 26, 28, 23, 27, 30, 24, 22, 31, 33])

# Interpretation based on correlation analysis
if np.corrcoef(daily_calorie_intake, bmi)[0, 1] > 0:
    print("Consuming more calories tends to be associated with a higher BMI.")
else:
    print("There is no significant positive correlation between calorie intake and BMI.")

if np.corrcoef(exercise_duration, bmi)[0, 1] < 0:
    print("Longer exercise duration tends to be associated with a lower BMI.")
else:
    print("There is no significant negative correlation between exercise duration and BMI.")
```
